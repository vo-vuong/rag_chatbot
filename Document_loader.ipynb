```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
# pdf_loader = PyPDFLoader('Stavian_Group.pdf', extract_images=True)
pdf_loader = PyPDFLoader('Stavian_Group.pdf')
docs = pdf_loader.load()
len(docs)
```

```python
docs[0]
```

```python
docs[1].page_content
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from typing import List

chunk_size = 300
chunk_overlap = 30
separator: List[str] = ['\n\n', '\n', ' ']

char_splitter = RecursiveCharacterTextSplitter(
    chunk_size=chunk_size,
    chunk_overlap=chunk_overlap,
    separators=separator,
    length_function=len,
    is_separator_regex=False
)
```

```python
text_out = char_splitter.split_text(docs[1].page_content)
len(text_out)
```

```python
text_out[0]
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
embedding = HuggingFaceEmbeddings()
```

```python
text = "This is a test sentence."
vector = embedding.embed_query(text)
```

```python
len(vector)
```

```python
chunks = char_splitter.split_documents(docs)
```

```python
chunks
```

```python
len(chunks)
```

```python
from langchain_qdrant import Qdrant
```

```python
## Add new docs and embedding to langchain_qdrant

qdrant_doc = Qdrant.from_documents(
    documents=chunks,
    embedding=embedding,
    url="http://localhost:6333",
    prefer_grpc=False,
    collection_name='test_collection'
    # api_key=_constants.QDRANT_API_KEY,
)
```

```python
query = "When: Khi làm việc và phục vụ khách hàng"
similar_docs = qdrant_doc.similarity_search(query, k=3)
```
